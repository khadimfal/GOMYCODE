```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Video games.csv')
df.dropna(inplace=True)
df = df[df['Global_Sales'] > 0]
plt.fill_between(df['Year'], df['Global_Sales'], alpha=0.5)
plt.plot(df['Year'], df['Global_Sales'])
plt.title('Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (millions)')
plt.show()

```

```python
plt.hist(df['Global_Sales'], bins=20)
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales (millions)')
plt.ylabel('Frequency')
plt.show()
```

```python
 plt.scatter(df['NA_Sales'], df['EU_Sales'])
plt.title('North American Sales vs European Sales')
plt.xlabel('North American Sales (millions)')
plt.ylabel('European Sales (millions)')
plt.show()
```

```python
plt.scatter(df['JP_Sales'], df['Other_Sales'])
plt.title('Japanese Sales vs Other Sales')
plt.xlabel('Japanese Sales (millions)')
plt.ylabel('Other Sales (millions)')
plt.show()
```

```python
platforms = ['Wii', 'PS3', 'X360', 'DS', 'PS2']
df_platforms = df[df['Platform'].isin(platforms)]
df_platforms = df_platforms.pivot_table(index='Year', columns='Platform', values='Global_Sales', aggfunc='sum')

df_platforms.plot(kind='area', stacked=True)
plt.title('Distribution of Sales by Platform Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (millions)')
plt.show()
```

```python
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(10)
top_publishers.plot(kind='bar')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (millions)')
plt.show()
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
plt.pie(genre_sales, labels=genre_sales
```
